```python
import pandas as pd
DataTable = pd.read_excel('Online Retail.xlsx')
DataTable.head()
```

```python
DataTable.info()
```

```python
DataTable['CustomerID'] = DataTable['CustomerID'].fillna(0).astype(int)
print(DataTable.dtypes)
```

```python
DataTable.head()
```

```python
Req_Data = DataTable[['InvoiceNo', 'StockCode', 'Description', 'Quantity','InvoiceDate','CustomerID']]
Req_Data.head()
```

```python
Req_Data.groupby(['CustomerID','Description','InvoiceDate']).agg(orders=('InvoiceNo', 'nunique'),quantity=('Quantity', 'sum')).sort_values(by='orders', ascending=False).head(10)
```

```python
Req_Item = Req_Data.pivot_table(index=('InvoiceNo','InvoiceDate','CustomerID'), columns=['Description'], values='Quantity').fillna(0)
Req_Item.head(3)
```

```python
def recommend_date(DataTable, CustomerID , InvoiceDate):
    return DataTable.loc[(DataTable['CustomerID'] == CustomerID) & (DataTable['InvoiceDate'] == InvoiceDate), 'Description']
```

```python
recommend_date(DataTable,17850,"2010-12-1 08:26:00") 
```

```python
def recommend_item(DataTable, item):
    recommendation = DataTable.corrwith(DataTable[item])
    recommendation.dropna(inplace=True)
    recommendation = pd.DataFrame(recommendation, columns=['correlation']).reset_index()
    recommendation = recommendation.sort_values(by='correlation', ascending=False)
    return recommendation
```

```python
recommendation = recommend_item(Req_Item, 'PARTY BUNTING')
recommendation.head()
```
